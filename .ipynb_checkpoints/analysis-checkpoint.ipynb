```python
def fill_decks(pclass, fare, cab):
    cab = str(cab)
    if pclass == 3 and cab == "nan":
        if fare <= 7.75:
            return 7
        if fare <= 8.05 and fare > 7.75:
            return 6
        if fare > 8.05 and fare <= 15.5:
            return 5
        else:
            return 4
    if pclass == 2 and cab == "nan":
        if fare <= 13:
            return 7
        if fare > 13 and fare <= 14.25:
            return 6
        if fare > 14.25 and fare <= 26:
            return 5
        else:
            return 4
    if pclass == 1 and cab == "nan":
        if fare <= 30.93:
            return 5
        if fare > 30.93  and fare <= 60.3:
            return 4
        if fare > 60.3 and fare <= 93.5:
            return 3
        else:
            return 2
    else:
        if cab[0] == "A":
            return 1
        if cab[0] == "B":
            return 2
        if cab[0] == "C":
            return 3
        if cab[0] == "D":
            return 4
        if cab[0] == "E":
            return 5
        if cab[0] == "F":
            return 6
        if cab[0] == "G":
            return 7
```

```python
train_df["Deck"] = train_df.apply(lambda x: fill_decks(x.Pclass, x.Fare, x.Cabin), axis=1)
test_df["Deck"] = test_df.apply(lambda x: fill_decks(x.Pclass, x.Fare, x.Cabin), axis=1)
```

```python
train_df.head()
```

```python
train_df.head()
```

```python
train_df.Embarked.fillna("S", inplace=True)
test_df.Embarked.fillna("S", inplace=True)
```

```python
train_df.describe()
```

```python
def titles(parch, sex):
    if sex == "male":
        return 1
    if parch == 0 and sex == "female":
        return 2
    else:
        return 3
```

```python
train_df["Title"] = train_df.apply(lambda x: titles(x.Parch, x.Sex), axis=1)
test_df["Title"] = train_df.apply(lambda x: titles(x.Parch, x.Sex), axis=1)
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
train_df.groupby(by="Title").Age.aggregate(func="mean")
```

```python
def fix_age(age, title):
    age = str(age)
    if age == "nan":
        if title == 1:
            return 30
        if title == 2:
            return 30
        if title == 3:
            return 24
    else:
        return age
```

```python
train_df.Age = train_df.apply(lambda x: fix_age(x.Age, x.Title), axis=1)
test_df.Age = test_df.apply(lambda x: fix_age(x.Age, x.Title), axis=1)
```

```python
train_df.Sex = train_df.Sex.map(lambda x: 1 if x == "male" else 0)
test_df.Sex = test_df.Sex.map(lambda x: 1 if x == "male" else 0)
```

```python
train_df.drop(labels="Cabin", inplace=True, axis=1)
```

```python
train_df.dropna(inplace=True)
```

```python
x_train = train_df[["Pclass", "Sex", "Age","SibSp","Parch", "Deck"]]
y_train = train_df.Survived
x_test = test_df[["Pclass", "Sex", "Age","SibSp","Parch",  "Deck"]]
x_test_ids = test_df.PassengerId

x_train.shape[0]
```

## End of preprocessing

In the sections above, I dealt with the NaN values, as well as did some feature engineering for the dataframe. In the end, I dropped columns that were not deemed necessary for the analysis, such as PassengerId and Name.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.kernel_approximation import Nystroem
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
n_feat = Nystroem(gamma=.1, random_state=12, n_components=200)
x_feat_train = n_feat.fit_transform(x_train)
x_feat_test = n_feat.transform(x_test)
scal = StandardScaler()
xtrain_st = scal.fit_transform(x_feat_train)
xtest_st = scal.transform(x_feat_test)
len(xtest_st[1])
```

```python
dist = []
for k in range(2,20):
    km = KMeans(random_state=12, n_clusters=k)
    clust_train = km.fit(xtrain_st)
    dist.append(km.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(range(2,20), dist)
plt.xlabel('Cluster amount')
plt.ylabel('Distortion')
plt.title("Elbow method")
plt.show()
```

```python
km = KMeans(random_state=12, n_clusters=16)
clust_train = km.fit_predict(xtrain_st, y_train)
clust_train = (clust_train - clust_train.mean()) / clust_train.std()
xtr_c_included = np.insert(xtrain_st, obj=200, values=clust_train, axis=1)
len(xtr_c_included[1])
```

```python
clust_test = km.predict(xtest_st)
clust_test = (clust_test - clust_test.mean()) / clust_test.std()
xte_c_included = np.insert(xtest_st, obj=200, values=clust_test, axis=1)
len(xte_c_included[1])
```

```python
clf = SGDClassifier(loss="modified_huber", random_state=12, penalty="elasticnet", l1_ratio=0.4, learning_rate="adaptive", eta0=0.00395, power_t=1)
clf.fit(xtr_c_included, y_train)
clf.score(xtr_c_included, y_train)
```

```python
res = clf.predict(xte_c_included)
```

```python
result = pd.DataFrame(res, columns=["Predictions"])
result["PassengerId"] = x_test_ids
```

```python
result
```

```python
result = pd.merge(result, fact, on="PassengerId", how="inner")
```
